### **Theory Question's**

**1. What is Simple Linear Regression**  
Simple Linear Regression is a method to model the relationship between one independent variable (X) and one dependent variable (Y) using a straight line: Y = mX + c.


**2. What are the key assumptions of Simple Linear Regression**  
- Linearity between X and Y  
- Independence of observations  
- Homoscedasticity (constant variance of errors)  
- Normal distribution of errors


**3. What does the coefficient m represent in the equation Y = mX + c**  
The coefficient **m** represents the **slope**, showing how much Y changes for a one-unit change in X.


**4. What does the intercept c represent in the equation Y = mX + c**  
The **intercept c** is the value of Y when X = 0. It shows the starting point of the line on the Y-axis.


**5. How do we calculate the slope m in Simple Linear Regression**  
\( m = \frac{ \sum (X_i - \bar{X})(Y_i - \bar{Y}) }{ \sum (X_i - \bar{X})^2 } \)


**6. What is the purpose of the least squares method in Simple Linear Regression**  
It minimizes the sum of squared differences between the actual Y values and predicted Y values.


**7. How is the coefficient of determination (R²) interpreted in Simple Linear Regression**  
R² tells how much of the variation in Y is explained by X. Ranges from 0 to 1 (higher is better).



**8. What is Multiple Linear Regression**  
A method to model the relationship between one dependent variable and two or more independent variables.


**9. What is the main difference between Simple and Multiple Linear Regression**  
Simple uses **one** independent variable; Multiple uses **two or more**.


**10. What are the key assumptions of Multiple Linear Regression**  
- Linearity  
- Independence  
- Homoscedasticity  
- Normality of errors  
- No multicollinearity


**11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model**  
It means the error variance is not constant. It can make standard errors unreliable, affecting confidence intervals and p-values.


**12. How can you improve a Multiple Linear Regression model with high multicollinearity**  
- Remove or combine correlated variables  
- Use regularization (like Ridge or Lasso)  
- Apply Principal Component Analysis (PCA)


**13. What are some common techniques for transforming categorical variables for use in regression models**  
- One-Hot Encoding  
- Label Encoding  
- Dummy Variables


**14. What is the role of interaction terms in Multiple Linear Regression**  
They show the combined effect of two variables on the outcome, beyond their individual effects.


**15. How can the interpretation of intercept differ between Simple and Multiple Linear Regression**  
In Simple, it's Y when X = 0. In Multiple, it's Y when **all X variables = 0**, which may or may not be meaningful.


**16. What is the significance of the slope in regression analysis, and how does it affect predictions**  
The slope tells how Y changes with a one-unit change in X. It directly affects the predicted value of Y.


**17. What are the limitations of using R² as a sole measure of model performance**  
- It always increases with more variables  
- Doesn’t show if predictors are significant  
- Doesn’t check assumptions or overfitting


**18. How would you interpret a large standard error for a regression coefficient**  
It suggests the estimate of the coefficient is not precise and may not be statistically significant.


**19. What is polynomial regression**  
A form of regression where the relationship between X and Y is modeled as an nth-degree polynomial.


**20. When is polynomial regression used**  
When data shows a **non-linear** relationship between the independent and dependent variables.


**21. How does the intercept in a regression model provide context for the relationship between variables**  
It gives the expected value of Y when all X variables are 0, setting the baseline for predictions.


**22. How can heteroscedasticity be identified in residual plots, and why is it important to address it**  
Look for a funnel shape in residual plots. It matters because it violates regression assumptions and affects reliability.


**23. What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²**  
It suggests that some variables may be unnecessary and not improving the model meaningfully.


**24. Why is it important to scale variables in Multiple Linear Regression**  
To ensure that variables are on the same scale, especially important when using regularization techniques.


**25. How does polynomial regression differ from linear regression**  
Linear regression models straight-line relationships; polynomial regression models curves using powers of X.


**26. What is the general equation for polynomial regression**  
Y = a + b₁X + b₂X² + b₃X³ + ... + bₙXⁿ


**27. Can polynomial regression be applied to multiple variables**  
Yes, it's called **multiple polynomial regression**, using polynomial terms of several variables.


**28. What are the limitations of polynomial regression**  
- Can overfit data  
- Sensitive to outliers  
- Interpretation becomes complex


**29. What methods can be used to evaluate model fit when selecting the degree of a polynomial**  
- Adjusted R²  
- Cross-validation  
- AIC/BIC  
- Residual plots


**30. Why is visualization important in polynomial regression**  
To understand the curve fit, detect overfitting, and ensure the model captures the data trend correctly.


**31. How is polynomial regression implemented in Python**  
Using `numpy.polyfit`, `sklearn.preprocessing.PolynomialFeatures`, and `LinearRegression` from `sklearn`.
