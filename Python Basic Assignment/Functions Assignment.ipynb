5. What are iterators in Python and how do they differ from iterables?
  - An iterable is any object in Python that can return an iterator, which means it can be traversed (looped over) using a for loop or other iteration methods.
Examples of iterables include lists, tuples, strings, dictionaries, sets, and more

  - An iterator is an object that represents a stream of data, and it keeps track of its state during iteration.
An iterator must implement two methods:
__iter__(): Returns the iterator object itself (usually self).
__next__(): Returns the next value in the sequence and advances the iterator. When there are no more items, it raises a StopIteration exception.

6. Explain the concept of generators in Python and how they are defined?
  - A generator is a special type of iterable in Python that allows you to iterate over a sequence of values lazily, meaning values are produced on the fly as needed, rather than all at once. This can save memory and improve performance, especially when working with large data sets or infinite sequences.
  - Generators are a type of iterable that produce items lazily, one at a time, using the yield keyword.
They are memory-efficient and useful for working with large datasets or streams of data.
A generator can be defined using a function with yield or using a compact generator expression.

9. Explain the purpose and usage of the `map()` function in Python?
  - The map() function in Python is used to apply a given function to all items in an iterable (like a list, tuple, etc.) and return an iterator that yields the results. It allows you to process all elements in an iterable without using an explicit loop.
  ```
  Example
numbers = [1, 2, 3, 4]
squared_numbers = map(lambda x: x ** 2, numbers)
#Convert to list and print the results
print(list(squared_numbers))  # Outputs: [1, 4, 9, 16]
```


10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
  - map(): Applies a function to each item, returning a transformed iterable.
  ```
  Example
  numbers = [1, 2, 3]
  squared = map(lambda x: x ** 2, numbers)
  print(list(squared))  # [1, 4, 9]
  ```
  - reduce(): Reduces an iterable to a single value by applying a cumulative function.
  ```
  Example
  from functools import reduce
  numbers = [1, 2, 3, 4]
  product = reduce(lambda x, y: x * y, numbers)
  print(product)  # 24
  ```
  - filter(): Filters an iterable based on a condition, returning items that satisfy it.
  ```
  Example
  numbers = [1, 2, 3, 4, 5]
  even_numbers = filter(lambda x: x % 2 == 0, numbers)
  print(list(even_numbers))  # [2, 4]
  ```

````python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

numbers = [12, 58, 52, 87, 9, 72]
result = sum_of_even_numbers(numbers)
print(result)
````

````python
#3.  Implement a Python function that takes a list of integers and returns a new list containing the squares of each number?

def square_num(numbers):
    return [num ** 2 for num in numbers]

input_list = [1, 2, 3, 4, 5]
squared_list = square_num(input_list)
print(squared_list)
````

````python
#4.  Write a Python function that checks if a given number is prime or not from 1 to 200?

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(num,end=",")
````

````python
#5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms?

class FibonacciIterator:
    def __init__(self, n):
        self.n = n
        self.a, self.b = 0, 1
        self.count = 0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n:
            result = self.a
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return result
        else:
            raise StopIteration

fibonacci = FibonacciIterator(10)
for num in fibonacci:
    print(num, end=",")
````

````python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

def powers_of_2(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for power in powers_of_2(5):
    print(power,end=",")
````

````python
#7. Implement a generator function that reads a file line by line and yields each line as a string?

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # Use strip() to remove any extra newline characters

# Example usage:
file_path = 'example.txt'  # Replace with your file path
for line in read_file_line_by_line(file_path):
    print(line)
````

````python
#8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple?

tuples_list = [(1, 3), (4, 1), (2, 2), (5, 4)]

sorted_list = sorted(tuples_list, key=lambda x: x[1])
print(sorted_list)
````

````python
#9. Create a Python program that uses `filter()` to remove all the vowels from a given string?

def remove_vowels(input_string):
    vowels = "aeiouAEIOU"

    result = ''.join(filter(lambda x: x not in vowels, input_string))
    return result

input_string = "Hello, World!"
output_string = remove_vowels(input_string)
print(output_string)
````
